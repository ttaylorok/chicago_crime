```python
import pandas as pd
import shapely
from shapely.geometry import asShape, asPoint
import numpy as np
import re

pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 1000)

df = pd.read_csv(r'C:\Users\comp\Documents\chicago_crime\Libraries_-_Locations__Hours_and_Contact_Information.csv')
df
```

```python
from geoalchemy2 import Geometry, WKTElement
import  sqlalchemy 

def parse_point(x):
    a = np.array(np.fromstring(re.sub('[()]', '', x), dtype=float, sep=','))
    b = asPoint(np.flip(a)) 
    c = WKTElement(b, srid=4326)
    return c

# parse lat/long to WKTElement
df['geom'] = df['LOCATION'].apply(lambda x : parse_point(x))

# create 
pg = open('db_credentials.txt')
db = pg.read()
pg.close()

# load into db
engine = sqlalchemy.create_engine(db)
df.to_sql('libraries', engine, if_exists='replace', index=False, dtype={'geom': Geometry('POINT', srid= 4326)})
```

```python
#housing.columns[2:]
def filter_census(df):
    for x in df.columns[2:]:
        if df[x].dtype == 'object':
            df[x] = pd.to_numeric(df[x].str.extract(r'([\d\.\d]+)', expand=False))
```

```python
filter_census(housing)
housing['GEO_ID_11'] = housing['GEO_ID'].apply(lambda x : x[-11:])
housing
```
